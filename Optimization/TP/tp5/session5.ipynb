## Practical session 5: Logistic regression with stochastic gradient descent

The goal of today's session is to implement a stochastic gradient descent algorithm, understand its properties, and use it to solve a logistic regression problem. Stochastic gradient descent is particularly useful when the gradient of the objective function is too costly to compute exactly but cheap unbiased estimators are available. 

### Subsampling gradient estimators in learning

Recall the logistic regression setting from Session 3, where we some *data*  $\mathcal{D}$ consisting of $m$ *examples* $\{d_i\}$; each example consisting of a *feature* vector $a_i\in\mathbb{R}^d$ and an *observation* $b_i\in \mathcal{O}$: $\mathcal{D} = \{[a_i,b_i]\}_{i=1..m}$. The observations are binary: $\mathcal{O} = \{-1 , +1 \}$, and the *Logistic loss* is used to form the following optimization problem
\begin{align*}
\min_{x\in\mathbb{R}^d } f(x) := \frac{1}{m}  \sum_{i=1}^m  \log( 1+\exp(-b_i \langle a_i,x \rangle) ) + \frac{\lambda_2}{2} \|x\|_2^2.
\end{align*}
where the last term is added as a regularization (of type $\ell_2$, aka Tikhonov) to prevent overfitting.

Evaluating $\nabla f$ requires scanning all the $m$ data-points. In some situations, $m$ can be very large, and computing $\nabla f$ becomes too costly. However, it is possible to consider only a small subset $B \subset \{ 1,\dots,m\}$ of data points at each iteration. To see that, one can decompose $f$ as
$$
f(x) = \frac{1}{m} \sum_{i=1}^m f_i(x).
$$
Alternatively, if $I$ is a uniform random variable on $\{ 1,\dots,m\}$, we can see that $f(x) = \mathbb{E}_{I}[f_I(x)]$, and that if $\hat{i}$ is a realization of $I$, then, $\mathbb{E}[\nabla f_{\hat{i}}(x)] = \nabla f(x)$ and $\nabla f_{\hat{i}}$ is very cheap to evaluate.

### Stochastic gradient descent

Suppose that for any $x \in \mathbb{R}^d$, we can construct a random variable $G(x)$ such that $G(x) \approx \nabla f(x)$ (in some sense), then stochastic gradient descent (SGD) updates the current iteration $x_k$ to $x_{k+1}$ by performing
$$
x_{k+1} = x_k - \tau_k G(x_k).
$$
Again, the choice of step-size is important. To illustrate the main ideas about the choice of step-size, let us assume that for any $x \in \mathbb{R}^d$, $\mathbb{E}[G(x)] = \nabla f(x)$ (unbiasedness) and that sufficient assumptions hold on the Lipschitz continuity of $f$ and on the variance of the estimators, then if $f$ is convex and the $\tau_k$ not too big, usual convergence rates are of the form
$$
\mathbb{E} \left[ f \left( \frac{\sum_{k=0}^K \tau_k x_k}{\sum_{k=0}^K \tau_k} \right) \right] - f(x_*) \leq \frac{\|x_0 - x_*\|^2}{2 \sum_{k=0}^K \tau_k} + \sigma^2 \frac{\sum_{k=0}^K \tau_k^2}{2 \sum_{k=0}^K \tau_k}.
$$
We can observe many things from the above:
- if $\tau_k \equiv \tau$, then the right-hand side does not converge to zero, we recover the sub-linear deterministic convergence rate plus a constant term from the variance of the estimator,
- if $\sum \tau_k^2 < \infty$ and $\sum \tau_k = \infty$, then the right-hand side goes to zero, motivating decreasing schedules of the form $\tau_k = 1 / (k+1)$,
- the convergence holds for the averaged iterate (the convergence of the last iterate $x_K$ is trickier).

```python
from algorithms import *  
from utils import *
import problem1 as pb
             
iterMax = 25000                     
x_init = np.zeros(pb.n)    
```

> Run the SGD algorithm for fixed and decreasing step-sizes, and compare the last iterations and the last averaged iterations. Consider first $\tau = 1/L$ for the schedule with fixed step size, and then consider $\tau = 1/ 4L$.

```python
x_decr, x_decr_tab, x_avg_decr, x_avg_decr_tab = SGD(pb.f, pb.grad_f_subsampling, x_init, 1/pb.L, 'decreasing', iterMax)
x_cons, x_cons_tab, x_avg_cons, x_avg_cons_tab = SGD(pb.f, pb.grad_f_subsampling, x_init, 1/pb.L, 'constant', iterMax)
for name, x in [('decr_last', x_decr), ('decr_avg', x_avg_decr), ('cons_last', x_cons), ('cons_avg', x_avg_cons)]:
    print(name, pb.f(x), pb.prediction_test(x, PRINT=False)[1])
```

```python
f_avg_decr = []
f_decr = []
for i in range(x_decr_tab.shape[0]):
    f_avg_decr.append( pb.f(x_avg_decr_tab[i])) 
    f_decr.append( pb.f(x_decr_tab[i])) 


f_avg_cons = []
f_cons = []
for i in range(x_cons_tab.shape[0]):
    f_avg_cons.append( pb.f(x_avg_cons_tab[i])) 
    f_cons.append( pb.f(x_cons_tab[i])) 


plt.figure()
plt.plot( f_avg_decr, color="magenta", linewidth=1.5, linestyle="--",label='averaged iterates, decreasing step-sizes')
plt.plot( f_decr, color="blue", linewidth=1.5, linestyle="--",label='last iterate, decreasing step-sizes')
plt.plot( f_avg_cons, color="magenta", linewidth=1.5, linestyle="-",label='averaged iterates, constant step-size')
plt.plot( f_cons, color="blue", linewidth=1.5, linestyle="-",label='last iterate, constant step-size')
# plt.ylim(0.3,0.5)

plt.grid(True)
plt.legend()
plt.show()
```
